```python
#import networkx as nx
import pandas as pd
from wordcloud import WordCloud

import matplotlib.pyplot as plt
```

```python
patents = pd.read_csv('/home/jovyan/query-results/Coffee_thursday_test_5ec90546-45f9-4b43-a5e1-1b20b2503b91.csv')

patents.head()
```

```python
plt.figure(figsize=[15,7.5])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
